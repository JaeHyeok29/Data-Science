```python
df_pop.rename(columns={'행정구역': 'District',
              '총 인구수': 'population'}, inplace=True)
df_pop
```

```python
df_pop.columns = 'district', 'population', 'household', 'pop_household', 'male', 'female'
df_pop
```

```python
df_pop.info()
```

```python
# 동작은 둘 다 똑같다
# df_pop.index = df_pop.district.to_list()
df_pop.index = df_pop['district']
df_pop
```

```python
df_pop.drop(columns={'district'}, inplace=True)
df_pop
```

```python
df_pop.info()
```

```python
df_pop.dtypes
```

```python
df_pop['population'] = df_pop['population'].str.replace(
    ',', '').str.replace('.', '').astype(int)
```

```python
df_pop['male_ratio'] = round(df_pop.male / df_pop.population, 4) * 100
df_pop
```

```python
df_pop['gender_ratio'] = round(df_pop.female / df_pop.male, 2)
df_pop
```

```python
df_pop['female_ratio'] = round(df_pop.female / df_pop.population, 4) * 100
```

```python
df_pop.drop('gender_ratio', axis=1, inplace=True)
df_pop
```

```python
df_pop.male.sum()
```

```python
# 전국 남성 인구 수를 열로 나타내면 좋을 것 같다
df_pop.loc['전국'] = [
    df_pop.population.sum(),
    df_pop.household.sum(),
    round(df_pop.pop_household.mean(), 2),
    df_pop.male.sum(),
    df_pop.female.sum(),
    0.0,
    0.0
]
```

```python
df_pop
```

```python
df_pop.loc['전국', 'male_ratio'] = round(
    sum(df_pop.male) / sum(df_pop.population), 4) * 100
df_pop
```

```python
df_pop.loc['전국', 'female_ratio'] = round(
    sum(df_pop.female) / sum(df_pop.population), 4) * 100
df_pop
```
