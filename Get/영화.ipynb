영화 타이틀을 제목과 연도로 분리하기


```python
import re
y = re.compile('\(\d+\)')
year = []
title = []

for i in df_boxoffice['영화']:
    m = y.search(i)
    g = m.group()
    title.append(i.replace(' '+g, ''))
    year.append(g.replace('(', '').replace(')', ''))
print(title)
print(year)
```

```python
df_boxoffice['MovieTitle'] = title
df_boxoffice['Year'] = year
df_boxoffice.drop(columns=['영화'], inplace=True)
df_boxoffice
```

인덱스, 칼럼 명 영어로 바꾸기


```python
df_boxoffice.index.name = 'rank'
df_boxoffice
```

```python
df_boxoffice.rename(columns={'세계 흥행 수익': 'WWBoxOffice', '북미 흥행 수익': 'DomesticBoxOffice',
                    '북미 %': 'DomesticPercentage', '제작비': 'Budget'}, inplace=True)
df_boxoffice
```

```python
df_boxoffice = df_boxoffice.reindex(
    ['MovieTitle', 'Year', 'WWBoxOffice', 'DomesticBoxOffice', 'DomesticPercentage', 'Budget'], axis=1)
df_boxoffice
```

타입을 int로


```python
df_boxoffice['WWBoxOffice'] = df_boxoffice['WWBoxOffice'].str.replace(
    '$', '').str.replace(',', '').astype(int)
df_boxoffice['DomesticBoxOffice'] = df_boxoffice['DomesticBoxOffice'].str.replace(
    '$', '').str.replace(',', '').astype(int)
df_boxoffice['DomesticPercentage'] = df_boxoffice['DomesticPercentage'].str.replace(
    '%', '').astype(float)
df_boxoffice['Budget'] = df_boxoffice['Budget'].str.replace(
    '$', '').str.replace(',', '').astype(int)
df_boxoffice['Year'] = df_boxoffice['Year'].astype(int)
```

총합 구하기


```python
df_boxoffice.loc['total'] = [
    'title',
    'year',
    df_boxoffice.WWBoxOffice.sum(),
    df_boxoffice.DomesticBoxOffice.sum(),
    round(df_boxoffice.DomesticPercentage.mean(), 2),
    df_boxoffice.Budget.sum()
]
df_boxoffice
```
